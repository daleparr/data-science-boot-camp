```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load e-commerce data
cultural_resonance_df = pd.read_csv('cultural_resonace.csv')

# Convert all columns to numeric, setting errors='coerce' to handle non-numeric strings
cultural_resonance_df = cultural_resonance_df.apply(pd.to_numeric, errors='coerce')

# You can choose to fill NaN values with an appropriate value or drop them
# For example, to fill with the mean:
cultural_resonance_df.fillna(cultural_resonance_df.mean(), inplace=True)

# Now you can create the correlation matrix and the heatmap
correlation_matrix = cultural_resonance_df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()

```
