```python
import pandas as pd
import matplotlib.pyplot as plt

# Load e-commerce data
cultural_resonance_df = pd.read_csv('cultural_resonace.csv')


plt.figure(figsize=(10, 10))  # Adjust the figure size as needed
cultural_resonance_df.hist()
plt.xticks(rotation=45, fontsize=6)  # Rotate x-axis labels and adjust font size
plt.yticks(fontsize=6)  # Adjust font size of y-axis labels
plt.tight_layout()  # Adjust subplot params so that the subplot(s) fits in to the figure area
plt.show()
```
