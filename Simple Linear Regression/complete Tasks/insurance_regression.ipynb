```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('insurance.csv')
print(df.head(10))
```

```python
# Scatter plot with age on the X-axis and charges on the Y-axis

plt.scatter(df['age'], df['charges'])
plt.xlabel('Age')
plt.ylabel('Insurance Charges')
plt.title('Scatter Plot of Age vs Insurance Charges')
plt.show()
```

```python
# Using LinearRegression to fit a model

X = df[['age']]  # Feature
y = df['charges']  # Target variable
model = LinearRegression()
model.fit(X, y)
```

```python
y_pred = model.predict(X)
```

```python
# Plotting the scatter plot with the best-fit line

plt.scatter(X, y, color='blue')
plt.plot(X, y_pred, color='red')
plt.xlabel('Age')
plt.ylabel('Insurance Charges')
plt.title('Scatter Plot with Best-Fit Line: Age vs Insurance Charges')
plt.show()
```
